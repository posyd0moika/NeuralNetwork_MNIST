```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
# Анализируем базу данных
(x_train, y_train), (x_test, y_test) = mnist.load_data() #  выгружаем базу данных
x_train = x_train / 255.0 # приводим ее к виду где цвет пикселя находиться от 0 до 1
x_test = x_test / 255.0
```

```python
summ = np.zeros(shape=(28, 28), dtype=float) # создаём массив 0 размерностью  28 на 28

for i in range(len(x_train)): # проходим в цикле базу данных
    summ += x_train[i] # и добавляем к массиву 0, те значения, которые содержались в базе данных

summ = summ / len(x_train) # дели наш получившийся массив на количество элементов, тем самым получаем, чисто вхождений на каждый пиксель(частотый анализ всех изображений)

plt.imshow(summ, cmap=plt.cm.binary) # выводим график
```
